## Imports des modules

```python
from scipy.misc import imread
import cv2

import matplotlib.pyplot as plt
%matplotlib inline

from helpers import show_images
import featuresourcer

import importlib
importlib.reload(featuresourcer)
FeatureSourcer = featuresourcer.FeatureSourcer

import warnings
warnings.filterwarnings('ignore')
```

## Initialisation param√®tres

```python
sourcer_params = {
  'color_model': 'ycrcb',              # hls, hsv, yuv, ycrcb
  'bounding_box_size': 64,             #
  'number_of_orientations': 12,        # 6 - 12
  'pixels_per_cell': 8,                # 8, 16
  'cells_per_block': 1,                # 1, 2
}
```

## Calcul des features HOG

```python
start_frame = imread("datasets/cropped/insect/Bee/Screenshot 2019-04-03 at 11.27.00.png")
shape = start_frame.shape[:][:]
rapport = max(shape) // 128
start_frame = cv2.resize(start_frame, (shape[0] // rapport, shape[1] // rapport))

f = sourcer.features(start_frame)
print("feature shape:", f.shape)

rgb_img, a_img, b_img, c_img = sourcer.visualize()
show_images([rgb_img, a_img, b_img, c_img], per_row = 4, per_col = 1, W = 10, H = 2)
```
